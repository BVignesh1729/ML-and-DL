```python
housing.dropna(axis=0, subset=['SalePrice'], inplace=True)
```

```python
housing.shape
```

```python
corr_matrix = housing.corr()
d = corr_matrix['SalePrice'].sort_values(ascending=False)

filt = (d>0.3) | (d < -0.1)
num_col_full = list(d[filt].index)[1:]
filt
```

```python
labels = housing['SalePrice']
housing.drop('SalePrice', axis=1, inplace=True)
housing
```

```python
cat_col_full = [col for col in housing.columns if housing[col].dtype=='object']
cat_col_full
```

```python
#one_hot_col = [col for col in cat_col_full if housing[col].nunique() < 10 and housing[col].isna().sum()<50]
#one_hot_col
```

```python
num_col_full
```

```python
print(len(cat_col_full), len(num_col_full))
```

```python
#Dealing with numerical coordinates
#Choosing to imputate
num_col_impute = [col for col in num_col_full if housing[col].isna().sum()<300]
num_col_impute
```

```python
num_col_drop = list(set(num_col_full) - set(num_col_impute))
```

```python
num_col_drop
```

```python
housing.drop(num_col_drop, axis=1, inplace=True)
```

```python
from sklearn.impute import SimpleImputer

impute = SimpleImputer(strategy='median')

impute_train_set = pd.DataFrame(impute.fit_transform(housing[num_col_impute]))

impute_train_set.columns = num_col_impute
impute_train_set.shape
```

```python
#Dealing with Categorical data
#CHoosing columns to drop if too many null
cat_col_include = [col for col in cat_col_full if housing[col].isna().sum() < 300]
```

```python
cat_col_include
```

```python
#Imputing for categorical columns

cat_impute = SimpleImputer(strategy='constant')

cat_data = pd.DataFrame(cat_impute.fit_transform(housing[cat_col_include]))
cat_data.columns = cat_col_include
```

```python
cat_data
```

```python
cat_col_ordinal = ['BsmtQual',
                    'BsmtCond',
                    'BsmtExposure',
                    'BsmtFinType1',
                    'BsmtFinType2',
                   'ExterQual',
                     'ExterCond',
                   'LotShape',
                     'LandContour',
                   'Utilities',
                   'LandSlope'
                    ]
```

```python
cat_col_one_hot = list(set(cat_col_include) - set(cat_col_ordinal))
```

```python
cat_col_one_hot
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value = -1)

cat_data_ord = pd.DataFrame(ordinal.fit_transform(cat_data[cat_col_ordinal]))
cat_data_ord.columns = cat_col_ordinal
```

```python
#One-Hot Encoding
from sklearn.preprocessing import OneHotEncoder

one_hot = OneHotEncoder(handle_unknown='ignore', sparse=False)
cat_data_oh = pd.DataFrame(one_hot.fit_transform(cat_data[cat_col_one_hot]))
#cat_data_oh.columns = cat_col_one_hot
cat_data_oh.shape
print(impute_train_set.shape, cat_data_ord.shape, cat_data_oh.shape)
final_train_set = pd.concat([impute_train_set, cat_data_ord, cat_data_oh], axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

rf_model = RandomForestRegressor(n_estimators = 100, random_state = 0)
scores = -1*cross_val_score(rf_model, final_train_set, labels, cv=5, scoring='neg_mean_absolute_error')

print(scores.mean())
```

```python
#Transforming the test_set for prediction(without setting column names)
test.drop(num_col_drop, axis=1, inplace=True)

impute_test_set = pd.DataFrame(impute.transform(test[num_col_impute]))
cat_data_test = pd.DataFrame(cat_impute.transform(test[cat_col_include]))
cat_data_test.columns = cat_col_include
cat_data_test_ord = pd.DataFrame(ordinal.transform(cat_data_test[cat_col_ordinal]))
cat_data_test_oh = pd.DataFrame(one_hot.transform(cat_data_test[cat_col_one_hot]))

final_test_set = pd.concat([impute_test_set, cat_data_test_ord, cat_data_test_oh], axis=1)
```

```python
rf_model.fit(final_train_set, labels)
prediction = rf_model.predict(final_test_set)
final_test_set.index = final_test_set.index + 1461 
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(n_estimators = 500, learning_rate = 0.01, n_jobs=5)
scores = -1*cross_val_score(xgb_model, final_train_set, labels, cv=5,scoring='neg_mean_absolute_error')
print(scores.mean())
```
