# Method 2

```python
import numpy as np
import pandas as pd
from sklearn import model_selection
```

```python
df_train = pd.read_csv('30-days-of-ml/train.csv')
```

```python
df_train.target.hist()
```

```python
df_train['kfold'] = -1
```

```python
df_train
```

```python
kf = model_selection.KFold(n_splits = 5, shuffle=True, random_state=42)
for fold, (train_indices, valid_indices) in enumerate(kf.split(X=df_train)):
    df_train.loc[valid_indices, 'kfold'] = fold
    
```

```python
df_train.head()
```

The idea here is that now every row in the dataframe has been assigned a fold number. Also, since the fold had been shuffled initially, we were able to assure in a certain way that each fold would be stratified(Check with histogram on the target for each fold for verification).

```python
import pandas
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
```

```python
df = pd.read_csv('train_folds.csv')
df_test = pd.read_csv('30-days-of-ml/test.csv')
```

```python
useful_features = [col for col in df.columns if col not in ('id', 'kfold', 'target')]
object_cols = [col for col in useful_features if 'cat' in col]
df_test = df_test[useful_features]
```

```python
final_predictions = []
```

```python
df.shape, df_test.shape
```

```python
for fold in range(5):
    xtrain = df[df.kfold != fold].reset_index(drop=True)
    xvalid = df[df.kfold == fold].reset_index(drop=True)
    xtest = df_test.copy()
    ytrain = xtrain['target']
    yvalid = xvalid['target']
    ordinal_encoder = OrdinalEncoder()
    
    xtrain[object_cols] = ordinal_encoder.fit_transform(xtrain[object_cols])
    xvalid[object_cols] = ordinal_encoder.transform(xvalid[object_cols])
    xtest[object_cols] = ordinal_encoder.transform(xtest[object_cols])
    
    model = XGBRegressor(random_state = fold, n_jobs=4)
    model.fit(xtrain[useful_features], ytrain)
    preds_valid = model.predict(xvalid[useful_features])
    test_preds = model.predict(xtest)
    
    final_predictions.append(test_preds)
    
    print(fold, mean_squared_error(yvalid, preds_valid, squared=False))
```

```python
p1, p2, p3, p4, p5 = final_predictions
```

## Method 3(with feature engineering) 

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
```

```python
df = pd.read_csv('train_folds.csv')
df_test = pd.read_csv('30-days-of-ml/test.csv')
df_test
```

```python
final_predictions = []
scores = []
useful_features = [col for col in df.columns if col not in ('id', 'kfold', 'target')]
object_cols = [col for col in useful_features if 'cat' in col]
numeric_cols = [col for col in useful_features if 'cont' in col]

for fold in range(5):
    xtrain = df[df.kfold != fold].reset_index(drop=True)
    xvalid = df[df.kfold == fold].reset_index(drop=True)
    xtest = df_test.copy()
    ytrain = xtrain['target']
    yvalid = xvalid['target']
    
    
#     print(xtrain.shape, xvalid.shape, xtest.shape)
    
    xtrain.drop(['id','target','kfold'], axis=1, inplace=True)
    xvalid.drop(['id','target','kfold'], axis=1, inplace=True)
    xtest.drop(['id'], axis=1,inplace=True)
    
    
#     print(xtrain.shape, xvalid.shape, xtest.shape)
    
    ohe = preprocessing.OneHotEncoder(sparse=False, handle_unknown='ignore')
    scaler = preprocessing.StandardScaler()
    
    xtrain_ohe = pd.DataFrame(ohe.fit_transform(xtrain[object_cols]))
    xvalid_ohe = pd.DataFrame(ohe.transform(xvalid[object_cols]))
    xtest_ohe = pd.DataFrame(ohe.transform(xtest[object_cols]))
    
    xtrain.drop(object_cols, axis=1, inplace=True)
    xvalid.drop(object_cols, axis=1, inplace=True)
    xtest.drop(object_cols, axis=1, inplace=True)
    
    xtrain = pd.concat([xtrain, xtrain_ohe], axis=1)
    xvalid = pd.concat([xvalid, xvalid_ohe], axis=1)
    xtest = pd.concat([xtest, xtest_ohe], axis=1)
    
#     print(xtrain.shape, xvalid.shape, xtest.shape)
    
    xtrain = scaler.fit_transform(xtrain)
    xvalid = scaler.transform(xvalid)
    xtest = scaler.transform(xtest)
    
    print(xtrain.shape, xvalid.shape, xtest.shape)
    
    model = XGBRegressor(random_state = fold, n_jobs=4)
    model.fit(xtrain, ytrain)
    preds_valid = model.predict(xvalid)
    test_preds = model.predict(xtest)
    
    final_predictions.append(test_preds)
    
    rmse = mean_squared_error(yvalid, preds_valid, squared=False)
    
    print(fold, rmse)
    scores.append(rmse)
    
print(np.mean(scores), np.std(scores))
```

```python
p1, p2, p3, p4, p5 = final_predictions
predictions = p1+p2+p3+p4+p5
predictions /= 5
df_test = pd.read_csv('30-days-of-ml/test.csv')
index = df_test.id
output = pd.DataFrame({'id': df_test.id,
                       'target': predictions})
output.to_csv('submission.csv', index=False)
```

# Feature engineering with target encoding
