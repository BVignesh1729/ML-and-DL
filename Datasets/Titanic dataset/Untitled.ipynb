```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('titanic/train.csv')
df_test = pd.read_csv('titanic/test.csv')
```

```python
df_train['Survived'].isnull().any()
```

```python
X = df_train.drop('Survived', axis=1)
y = df_train['Survived']
X_test = df_test.copy()
X
```

# Imputation

```python
X.isnull().sum()
```

```python
#For training set
X.drop('Cabin', axis=1, inplace=True)

#For imputing age
Age_data = X[['Pclass', 'Sex']]
Age_label = X['Age']
```

```python
filt = Age_label.isnull()
Age_label[~filt]
AX_train = Age_data[~filt]
AX_test = Age_data[filt]
Ay_train = Age_label[~filt]
AX_train
```

```python
AX_test['Pclass'].value_counts()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OrdinalEncoder

ordinal = OrdinalEncoder()
ordinal.fit(AX_train)
AX_train = pd.DataFrame(ordinal.transform(AX_train))
AX_test = pd.DataFrame(ordinal.transform(AX_test))
model = RandomForestRegressor(n_estimators=25, random_state=42)
model.fit(AX_train, Ay_train)
Ay_predict = model.predict(AX_test)
```
