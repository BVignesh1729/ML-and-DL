# Preprocessing

```python
#Combining dataframes
movies = movies.merge(credits,on='title')
```

```python
# Required columns(for creating tags)
# Human engineered analysis

cols = ['id', 'genres', 'keywords', 'title', 'overview', 'popularity', 'release_date', 'revenue', 'cast', 'crew']

num_col = [col for col in cols if movies[col].dtype != 'object']
text_col = list(set(cols) - set(num_col))
```

```python
movies['original_language'].value_counts()
```

```python
data = movies[cols].copy()
```

```python
#Data preprocessing
#Removing missing data

data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.iloc[0]['genres']
```

```python
import ast
```

```python
#Helper function to convert the genre list of dicts to just list of words
def convert(obj):
    G = []
    for item in ast.literal_eval(obj):
        G.append(item['name'])
    return G
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Applying it to the genres column

data['genres'] = data['genres'].apply(convert)
```

```python
data.head()
```

```python
#Applying the same for keywords
data['keywords'] = data['keywords'].apply(convert)
```

```python
data.head()
```

```python
#Helper function for cast(top 5 actors)
def top_5(obj):
    G = []
    cast = 0
    for item in ast.literal_eval(obj):
        if cast<5:
            G.append(item['name'])
            cast += 1
        else:
            break
    return G
```

```python
data['cast'] = data['cast'].apply(top_5)
```

```python
data.head()
```

```python
#Helper function to get the director name alone from crew
def director(obj):
    G = []
    for item in ast.literal_eval(obj):
        if item['job'] == 'Director':
            G.append(item['name'])
            break
    return G 
```

```python
data['crew'] = data['crew'].apply(director)
```

```python
data.head()
```

```python
#Converting overview to list

data['overview'] = data['overview'].apply(lambda x: x.split())
```

```python
data.head()
```

```python
#Removing spaces so that similarity is not confused for different entities with just few samw words
data[['genres', 'keywords', 'cast', 'crew']] = data[['genres', 'keywords', 'cast', 'crew']].applymap(lambda x:[i.replace(" ","") for i in x])
```

```python
data
```

```python
#Changing release data to a list with a string, so that we can concatenate with others to form the tag
def rel_date(obj):
    G = []
    G.append(obj)
    return G

data['release_date'] = data['release_date'].apply(rel_date)
```

```python
#Conocatenating
data['tags'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew'] + data['release_date']
```

```python
data
```

```python
movie_df = data[['id', 'title', 'tags']]
```

```python
#Converting back to space seperated string with lower case letters, to avoid dissimilarity between case sensitive words
movie_df['tags'] = movie_df['tags'].apply(lambda x: (" ".join(x)).lower())
```

```python
movie_df
```

```python
#Stemming(to remove other tenses of the same word, to remove plurals)
import nltk 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(txt):
    S = []
    for i in txt.split():
        S.append(ps.stem(i))
        
    return " ".join(S)
    
```

```python
movie_df['tags'] = movie_df['tags'].apply(stemming)
```

# Text Vectorization

```python
#Using count vectorizer
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=6000, stop_words='english')
```

```python
cv.fit(movie_df['tags'])
freq = cv.transform(movie_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
freq.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(freq)
```

```python
print(sim.shape)
print(sim)
```

```python
def recommend(movie):
    movie_index = movie_df[movie_df['title'] == movie].index[0]
    movie_rec = sorted(enumerate(sim[movie_index]),reverse=True,key=lambda x:x[1])[1:10]
    
    for mov in movie_rec:
        print(movie_df.iloc[mov[0]]['title'])
    
```

```python
recommend('Batman Begins')
```
