```python
def rv(value_list):
    return np.array([value_list])

def cv(value_list):
    return np.transpose(rv(value_list))

def f1(x):
    return float((2 * x + 3)**2)

def df1(x):
    return 2 * 2 * (2 * x + 3)

def f2(v):
    x = float(v[0]); y = float(v[1])
    return (x - 2.) * (x - 3.) * (x + 3.) * (x + 1.) + (x + y -1)**2

def df2(v):
    x = float(v[0]); y = float(v[1])
    return cv([(-3. + x) * (-2. + x) * (1. + x) + \
               (-3. + x) * (-2. + x) * (3. + x) + \
               (-3. + x) * (1. + x) * (3. + x) + \
               (-2. + x) * (1. + x) * (3. + x) + \
               2 * (-1. + x + y),
               2 * (-1. + x + y)])
```

```python
def num_grad(f, delta=0.001):
    def df(x):
        l = []
        d = x.shape[0]
        for i in range(d):
            grad = np.zeros(x.shape)
            grad[i,0] = delta
            k = (f(x+grad) - f(x-grad))/(2*delta)
            l.append(k)
        print(cv(l))
        return cv(l)
    return df
```

```python
def gd2(f, x0, step_size_fn, max_iter):
    theta = x0.copy()
    fs = []
    xs = []
    Df = num_grad(f)
    fs.append(f(theta))
    xs.append((theta))
    for i in range(max_iter):
        theta = theta - (step_size_fn(i))*(Df(theta))
        fs.append(f(theta))
        xs.append((theta))
    
    return theta, fs, xs 
```

```python
def gd3(f, df, x0, step_size_fn, max_iter):
    theta = x0.copy()
    fs = []
    xs = []
    fs.append(f(theta))
    xs.append(theta)
    for i in range(max_iter):
        theta = theta - (step_size_fn(i))*df(theta)
        fs.append(f(theta))
        xs.append(theta)
    
    return theta, fs, xs    
```

```python
def gd(f, df, x0, step_size_fn, max_iter):
    prev_x = x0
    fs = []; xs = []
    for i in range(max_iter):
        prev_f, prev_grad = f(prev_x), df(prev_x)
        fs.append(prev_f); xs.append(prev_x)
        if i == max_iter-1:
            return prev_x, fs, xs
        step = step_size_fn(i)
        prev_x = prev_x - step * prev_grad
```

```python
def package_ans(gd_vals):
    x, fs, xs = gd_vals
    return [x.tolist(), [fs[0], fs[-1]], [xs[0].tolist(), xs[-1].tolist()]]
```

```python
def hinge(v):
    return np.where(v<1, 1-v, 0)
```

```python
def hinge_loss(x, y, th, th0):
    return hinge(y*(np.dot(th.T,x) + th0))
```

```python
def svm_obj(x, y, th, th0, lam):
    d, n = x.shape
    print(x.shape)
    loss = 0
    for i in range(n):
        loss = loss + hinge_loss(x[:,i],y[0,i],th,th0)
    loss = loss/n
    loss += lam*(np.sum(th**2))
    return loss[0,0] #returning as a saclar
```
